# Adding title

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
phool =sns.load_dataset("iris")
phool
plt.title("phool ka data")
#draw a lineplot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.show()
```

# Adding limit

```python
import seaborn as sns
import matplotlib.pyplot as plt
phool =sns.load_dataset("iris")
phool
plt.title("phool ka data")
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.xlim(2)
plt.ylim(1)
plt.show()
```

# size of figure

```python
import seaborn as sns
import matplotlib.pyplot as plt
phool =sns.load_dataset("iris")
phool
plt.title("phool ka data")
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.figure(figsize=(4,5))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
phool =sns.load_dataset("iris")


plt.title("phool ka data")
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
sns.set_style("dark")
sns.set_style(style=None,rc=None)
plt.figure(figsize=(4,5))
plt.show()
```
