```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
boat =sns.load_dataset("titanic")
boat
sns.barplot(x="sex",y="alone",hue="survived",data=boat,order=["female","male"]#order of a data
plt.show()
```

### Estimator

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy
from numpy import mean
#load dataset
boat =sns.load_dataset("titanic")
boat
sns.barplot(x="sex",y="fare",hue="survived",data=boat, estimator=mean)#estimator mean
plt.show()
```

### saturation


```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy
from numpy import mean
#load dataset
boat =sns.load_dataset("titanic")
boat
sns.barplot(x="sex",y="fare",hue="survived",data=boat, estimator=mean, saturation=0.2)#estimator mean
plt.show()
```
