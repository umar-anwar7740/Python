```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
kashti=sns.load_dataset("titanic")
kashti
```

```python
sns.boxplot(x="sex",y="fare",data=kashti)
```

```python
sns.boxplot(x="survived",
            y="age",showmeans=True,
            meanprops={"marker":"+",
                      "markersize":"12",
                      "markeredgecolor":"red"},
            data=kashti)
#show labels
plt.xlabel("How many survived"),
plt.ylabel("Age"),
plt.title("Box plot of")
plt.show()
```
